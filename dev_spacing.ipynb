```python
import pandas as pd
from pykospacing import Spacing
```

## Data Load

```python
dev = pd.read_csv('dev.csv')
dev
```

## 띄어쓰기 교정 모델 Load

```python
spacing = Spacing()
```

```python
dev['sentence_1_space'] = dev['sentence_1'].apply(lambda x: spacing(x))
dev['sentence_2_space'] = dev['sentence_2'].apply(lambda x: spacing(x))
dev
```

### Token 개수 2개 이하인 행만 sapcing

```python
dev_1 = dev[dev['sentence_1'].apply(lambda x: len(x.split())) <= 2]
dev_1 = dev_1[dev_1['sentence_1'] != dev_1['sentence_1_space']]
dev_1 = dev_1[~dev_1.index.isin([230, 267, 447])]
dev_1 = dev_1[['sentence_1', 'sentence_1_space']]
dev_1 = pd.DataFrame(dev_1)
dev_1.to_csv('dev_1.csv')
dev_1
```

```python
dev_2 = dev[dev['sentence_2'].apply(lambda x: len(x.split())) <= 2]
dev_2 = dev_2[dev_2['sentence_2'] != dev_2['sentence_2_space']]
dev_2 = dev_2[~dev_2.index.isin([54, 215, 230, 361, 435, 491, 521])]
dev_2 = dev_2[['sentence_2', 'sentence_2_space']]
dev_2 = pd.DataFrame(dev_2)
dev_2.to_csv('dev_2.csv')
dev_2
```
